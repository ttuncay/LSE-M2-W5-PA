```python
#Import necessary libraries
import pandas as pd
```

```python
# Importing the excel data
transport_cost=pd.read_excel('/Users/tunatuncay/Documents/Personal/LSE Data Analytics/Module 2/Week 5/LSE_DA201_Week_5_files/Data/transport_costs.xlsx')
print(transport_cost.shape)
print(transport_cost.dtypes)
transport_cost.head()
```

```python
# Delete the columns currency and distance unit
transport_cost=transport_cost[['region', 'country', 'port', 'sea freight cost', 'road transport cost per km']]
transport_cost.head()
```

```python
# Calculating values for new variables
transport_cost['total_sea_freight_costs']=transport_cost['sea freight cost']*2*4
transport_cost['total_road_transport_cost']=transport_cost['road transport cost per km']*10000
transport_cost['total_export_costs']=transport_cost['total_sea_freight_costs']+transport_cost['total_road_transport_cost']

# Reordering the columns
transport_cost=transport_cost[['region', 'country', 'port', 'sea freight cost', 'total_sea_freight_costs', 'road transport cost per km', 'total_road_transport_cost', 'total_export_costs']]
transport_cost.head()                         

# Saving to csv file
transport_cost.to_csv('transport_cost_python.csv')
```

```python
# Showing the summary data as per region
cost_per_region=transport_cost.groupby('region')[['total_sea_freight_costs', 'total_road_transport_cost', 'total_export_costs']].sum()
cost_per_region
```

```python
# Saving to csv file
cost_per_region.to_csv('transport_cost_per_region_python.csv')
```
