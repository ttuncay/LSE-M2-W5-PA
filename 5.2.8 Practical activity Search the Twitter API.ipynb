## Search trends worldwide for the IDs of the required cities (US as New York and UK as London).


```python
# Identify trends worldwide
trends_worldwide=twitter_api.trends.available()

# Check how many trends available
print(len(trends_worldwide))
```

```python
# Getting woeid information of NY
list_of_NY=[_ for _ in trends_worldwide if _['name']=='New York']
print(len(list_of_NY))
print(list_of_NY[0])
```

```python
# Getting woeid information of London
list_of_London=[_ for _ in trends_worldwide if _['name']=='London']
print(len(list_of_London))
print(list_of_London[0])
```

```python
# Creating a subset with more than 50k tweet
ny_50ktrends_df=ny_trends_df[ny_trends_df['tweet_volume']>50000]
ny_50ktrends_df
```

## Search Twitter for #Bitcoin

```python
# Setting a variable to search
q='#Bitcoin'

# Set count to 100.
Count=100
```

```python
# Bringing tweets
search_results=twitter_api.search.tweets(q=q, count=100)
statuses=search_results['statuses']
```

```python
print(f"Length of the statutes is {len(statuses)}")

for _ in range(5):
    try:
        next_results= search_results['search_metadata']['next_results']
    except Keyerror:
        break
        
        # Create dict for results
        kwargs = dict([kv.split('=') for kv in next_results[1:].split('&')])
        
        search_results = twitter_api.search.tweets(**kwargs)
        statuses+=search_results['statuses']
  
print(json.dumps(statuses[0], indent=4))
# Print a sample one (the first one)
```
